# Ejercicio 10: Álgebra lineal con NumPy

```python
import numpy as np # Importamos la librería NumPy para operaciones con arrays y funciones de álgebra lineal

A = np.array([[3,2], [1,2]]) # Define la matriz de coeficientes A del sistema
b = np.array([5,5]) # Define el vector de términos independientes b

x = np.linalg.solve(A, b) # Resuelve el sistema de ecuaciones lineales Ax = b

print(x) # Muestra el vector solución encontrado
```
