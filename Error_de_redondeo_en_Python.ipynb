# Ejercicio 7: Error de redondeo en Python

```python
a = 0.1 + 0.2 # Suma dos n√∫meros decimales
b = 0.3 # Asignamos directamente el valor 0.3 a la variable b

print(a) # Muestra el valor almacenado en a
print(a == b) # Compara si a y b son exactamente iguales
print(abs(a - b)) # Calcula y muestra el valor absoluto de la diferencia entre a y b
```
