# Ejercicio 2: Error numérico en una aproximación

```python
import math # Proporciona funciones matemáticas, incluyendo sqrt() para calcular raíces cuadradas
```

```python
real = math.sqrt(2) # Calcula el valor de sqrt(2) usando la función math.sqrt() y lo almacena en la variable real
aprox = 1.41 # Define un valor aproximado de sqrt(2)

error_absoluto = abs(real - aprox) # Calcula el error absoluto
error_relativo = error_absoluto / real # Calcula el error relativo
print(error_absoluto, error_relativo) # Muestra ambos valores calculados
```
