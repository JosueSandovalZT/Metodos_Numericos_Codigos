# Ejercicio 5: Precisión vs Exactitud

```python
import numpy as np # Importamos la librería NumPy para operaciones numéricas
```

```python
mediciones = np.array([9.8, 9.81, 9.79, 9.80]) # Crea un array de NumPy que contiene 4 mediciones experimentales
valor_real = 9.81 # Define el valor de referencia o valor real

media = np.mean(mediciones) # Calcula la media aritmética (promedio) de las mediciones usando la función mean() de NumPy
error = abs(media - valor_real) # Calcula el error absoluto

print("Media:", media) # Muestra el valor de la media calculada
print("Error absoluto:", error) # Muestra el error absoluto calculado
```
