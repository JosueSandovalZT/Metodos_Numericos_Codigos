# Ejercicio 6: Sesgo numérico en datos

```python
import numpy as np # Importamos la librería NumPy para operaciones numéricas
```

```python
datos_reales = np.random.normal(0, 1, 1000) # Genera 1000 muestras aleatorias de una distribución normal (gaussiana)
sesgo = 0.5 # Define un valor de sesgo de 0.5 que se añadirá a los datos

datos_sesgados = datos_reales + sesgo # Crea un nuevo conjunto de datos añadiendo el sesgo a cada elemento

print("Media real:", np.mean(datos_reales)) # Calcula y muestra la media de los datos originales
print("Media sesgada:", np.mean(datos_sesgados)) # Calcula y muestra la media de los datos con sesgo
```
