# Ejercicio 11: Optimización numerica con SciPy

```python
from scipy.optimize import minimize # Importa la función minimize del submódulo optimize de la librería SciPy

f = lambda x: (x - 2) ** 2 # Define la función objetivo a minimizar

resultado = minimize(f, x0 = 0) # Ejecuta el algoritmo de optimización

print(resultado.x) # Muestra el punto donde se encontró el mínimo
```
