# Ejercicio 8: Error de truncamiento

```python
import math # Importa el módulo math que proporciona funciones matemáticas

x = 1 # Define el punto de evaluación de la función exponencial
aprox = 1 + x + (x ** 2) / 2 # Calcula una aproximación de e ** x usando los primeros tres términos de la serie de Taylor
real = math.exp(x) # Calcula el valor real de e ** x usando la función exp() del módulo math

error = abs(real - aprox) # Calcula el error absoluto entre el valor real y la aproximación
print("Aprox:", aprox) # Muestra el valor aproximado
print("Real:", real) # Muestra el valor real
print("Error:", error) # Muestra la diferencia absoluta
```
