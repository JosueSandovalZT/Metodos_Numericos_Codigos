# Ejercicio 12: Visualización del error

```python
import numpy as np # Importamos la librería NumPy para generar datos numéricos (arrays, operaciones matemáticas)
import matplotlib.pyplot as plt # Importamos el submódulo pyplot de Matplotlib para crear visualizaciones y gráficos

X = np.linspace(-1, 5, 100) # Crea un array con 100 puntos igualmente espaciados en el intervalo [-1, 5]
y = (X - 2) ** 2 # Calcula los valores de la función para cada punto en X

plt.plot(X, y) # Crea el gráfico de línea
plt.xlabel("X") # Añade una etiqueta al eje X
plt.ylabel("f(x)") # Añade una etiqueta al eje Y
plt.title("Función de costo") # Añade un título al gráfico
plt.show() # Muestra el gráfico en pantalla
```
