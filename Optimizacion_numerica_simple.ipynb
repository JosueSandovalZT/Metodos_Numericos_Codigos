# Ejercicio 3: Optimización numérica simple

```python
import numpy as np # Importamos la librería NumPy para operaciones numéricas
import math # Proporciona funciones matemáticas
```

```python
x = 0 # Inicializa la variable x en 0, que será el punto de partida para la optimización
alpha = 0.1 # Define la tasa de aprendizaje

for _ in range(20): # Bucle que se ejecutará 20 veces
    grad = 2 * (x - 3) # Calcula el gradiente (derivada) de la función
    x = x - alpha * grad # Actualiza el valor de x usando la regla del descenso de gradiente
print(x) # Muestra el valor final de x
```
