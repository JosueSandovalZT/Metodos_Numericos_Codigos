# Ejercicio 9: Propagación del error

```python
x = 2 # Define el valor de la variable independiente x
dx = 0.01 # Define la incertidumbre o error asociado a la medición de x

y = x ** 2 # Calcula el valor de la función
dy = abs (2 * x) * dx # Calcula la propagación del error usando derivadas

print("y:", y) # Muestra el valor calculado de la función
print("Error aproximado en y:", dy) # Muestra el error propagado estimado
```
